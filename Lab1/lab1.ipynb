## 3.Matplotlib (створити набір даних лінійної функції з урахуванням помилок вимірювань, побудувати графік функції та згенерованих даних, обчислити похибку з використанням метрик MAE, MSE). Записати отримані результати у csv файл (формат: X,Y, Y_hat, mAE, mSE)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl

%matplotlib inline

market = mpl.subplots()

def xy_array(start, count, step, function):
    x = []
    y = []
    stop = start + count * step

    for i in np.arange(start, stop, step):
        x.append(i)
        y.append(function(i))
    
    return np.array(x), np.array(y)

def parabola(x):
    return x ** 2

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)

def tan(x):
    return np.tanh(x)


def mae(actual, prediction):
    return np.absolute((actual - prediction))    

def mse(actual, prediction):
    return np.square(actual - prediction)


def yhat(x, y):
    b1, b0 = np.polyfit(x, y, 1)
    y_hat = b0 + b1*x
    return y_hat


x, y = xy_array(0, 200, 0.1, sin)
mpl.plot(x, y)
```

```python
x = np.random.randint(0, 200, 10)
y = np.random.randint(0, 200, 10)
mpl.scatter(x,y)
```

```python
figure = mpl.figure(figsize=(10, 10))
axis3 = figure.add_subplot(333, projection='3d')
axis3.scatter(x, y)
```

```python
x, y = xy_array(-100, 200, 1, parabola)
mpl.plot(x, y)
```

```python
figure, axis = mpl.subplots()

x1, y1 = xy_array(0, 200, 0.2, cos)
line1, = axis.plot(x1, y1)

x2, y2 = xy_array(0, 200, 0.1, sin)
line2, = axis.plot(x2, y2)

x3, y3 = xy_array(0, 200, 0.2, tan)
line3, = axis.plot(x3, y3)

x0, y0 = xy_array(0, 200, 0.2, lambda x: x / 15)
line0, = axis.plot(x0, y0)
```

```python
sin_yhat = yhat(y1, y0)
cos_yhat = yhat(y2, y0)
tan_yhat = yhat(y3, y1)


print('sin(x):', sin_yhat.mean())
print('cos(x):', cos_yhat.mean())
print('tanh(x):', tan_yhat.mean())
print()

fig, ax = mpl.subplots()

ax.plot(x1, sin_yhat)
ax.plot(x2, cos_yhat)
ax.plot(x3, tan_yhat)
```

```python
sin_mae = mae(y1, y0)
cos_mae = mae(y1, y0)
tan_mae = mae(y3, y0)


print('sin(x):', sin_mae.mean())
print('cos(x):', cos_mae.mean())
print('tanh(x):', tan_mae.mean())
print()


fig, ax = mpl.subplots()

ax.plot(x1, sin_mae)
ax.plot(x2, cos_mae)
ax.plot(x3, tan_mae)
```

```python
mse_sin = mse(y1, y0)
mse_cos = mse(y2, y0)
mse_tan = mse(y3, y0)


print('sin(x):', mse_sin.mean())
print('cos(x):', mse_cos.mean())
print('tanh(x):', mse_tan.mean())
print()

fig, ax = mpl.subplots()

ax.plot(x1, mse_sin)
ax.plot(x2, mse_cos)
ax.plot(x3, mse_tan)
```

```python
sin_arr = (x1, y1, sin_yhat, sin_mae, mse_sin)
cos_arr = (x2, y2, cos_yhat, cos_mae, mse_cos)
tan_arr = (x3, y3, tan_yhat, tan_mae, mse_tan)

data = np.column_stack((sin_arr, cos_arr, tan_arr))
data = np.transpose(data, (1, 0))

dframe = pd.DataFrame(data, columns=('X', 'Y', 'Y_hat', 'mAE', 'mSE'))
dframe.to_csv('table3.csv')
```
